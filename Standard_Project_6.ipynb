```python
from datetime import datetime as dt  #so ovoj poseben kod simnuvame ESG score od microsoft na mesecno nivo
import urllib.request
import json
dataframes = []

url = "https://query2.finance.yahoo.com/v1/finance/esgChart?symbol=MSFT"  #linkot od kadeshto mozhe da simneme

connection = urllib.request.urlopen(url)

data = connection.read()
data_2 = json.loads(data)
Formatdata = data_2["esgChart"]["result"][0]["symbolSeries"]
Esg_score_monthly = pd.DataFrame(Formatdata) #pravime format na datata so pandas
Esg_score_monthly
Esg_score_monthly["timestamp"] = pd.to_datetime(Esg_score_monthly["timestamp"], unit="s") #od string datumi vo vistinski datumi format gi menuvame



# do ovoj datum sakame da ni bide mnozhestvoto
cutoff_date = pd.to_datetime('2018-09-17')

# Transpose the DataFrame to work with columns
Esg_score_monthly = Esg_score_monthly.transpose()

# Identify columns where the 'date' is higher than the cutoff date -> ako ni e povekje od toj datum da se dropne
columns_to_drop = Esg_score_monthly.columns[Esg_score_monthly.loc['timestamp'] < cutoff_date]

# Drop the identified columns from the DataFrame (transposed)
Esg_score_monthly = Esg_score_monthly.drop(columns=columns_to_drop)

# Transpose the DataFrame back to the original format
Esg_score_monthly = Esg_score_monthly.transpose()
Esg_score_monthly.reset_index(drop=True, inplace=True)
Esg_score_monthly = Esg_score_monthly.infer_objects()  # Ensure columns have appropriate data types

Esg_score_monthly #datasetot na mesecno nivo, no falat nekoi podatoci
```

```python
#funkcija koja kje sluzhi za da predviduvame so xgb regressor. Ja pravime funkcijata za da predviduvame na 3te  datasets
#za da ne go pishuvame kodot 3 pati. Chitame dataset delima na X,Y trenirame, barame error i R^2
def XGB_reggresor(dataset):
  y=dataset['Open']
  dropcolumns=['Open','Date']
  x = dataset.drop(dropcolumns, axis=1)
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
  model = XGBRegressor()
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)

  mse = mean_squared_error(y_test, y_pred)


  rmse = np.sqrt(mse)


  mae = mean_absolute_error(y_test, y_pred)

  r_squared = r2_score(y_test, y_pred)

  print(f"Mean Squared Error: {mse}")
  print(f"Root Mean Squared Error: {rmse}")
  print(f"Mean Absolute Error: {mae}")
  print(f"R-squared: {r_squared}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

```

```python
#isto kako prethodnata funkcija no sega samo za regresija
def lin_regr(dataset):
  y=dataset['Open']
  dropcolumns=['Open','Date']
  x = dataset.drop(dropcolumns, axis=1)
  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
  model = LinearRegression()
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)
  rmse = mse**0.5
  r_squared = r2_score(y_test, y_pred)
  print(f"RMSE: {rmse}")
  print(f"R-squared: {r_squared}")
```
