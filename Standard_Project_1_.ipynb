```python
pearson_correlation = merged_data.corr(method='pearson')
print(pearson_correlation)
pearson_correlation
```

```python
data_subset = merged_data.drop('Date', axis=1)
print(data_subset)
window_size = 30
rolling_corr = data_subset.rolling(window_size).corr().dropna()
rolling_corr
```

```python
average_corr = rolling_corr.groupby(level=0).mean()
average_corr
```

```python
#cons: Not as sensitive to linear relationships as Pearson correlation
#pros: Robust to non-linearity
spearman_corr = merged_data.corr(method='spearman')
spearman_corr
```

```python
#cons: Not as sensitive to linear relationships as Pearson or Spearman correlation
#pros: Robust to non-linearity
kendall_corr = merged_data.corr(method='kendall')
kendall_corr
```

```python
data_subset = merged_data.drop('Date', axis=1)
print(data_subset)
window_size = 7
rolling_corr2 = data_subset.rolling(window_size).corr().dropna()
rolling_corr2

average_corr = rolling_corr2.mean(axis=1)

lowest_corr = rolling_corr2.min(axis=1)

highest_corr = rolling_corr2.max(axis=1)
highest_corr = highest_corr.where(highest_corr < 0.99999999999).dropna()
second_highest_corr = highest_corr.max()


print('\nSecond Highest Correlation 7d (Excluding 1.0):')
print(second_highest_corr)

print('\nAverage Correlation 7d:')
print(average_corr)

print('\nLowest Correlation 7d:')
print(lowest_corr)
```
