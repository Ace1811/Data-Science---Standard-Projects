```python


def is_all_numeric(df):
  """
  Check if all rows in a Pandas DataFrame are numeric.

  Args:
    df: The Pandas DataFrame to check.

  Returns:
    True if all rows are numeric, False otherwise.
  """

  for column in df.columns:
    if not pd.api.types.is_numeric_dtype(df[column]):
      return False

  return True


print(is_all_numeric(data))
```

```python

def find_non_numeric_rows(df):
  """
  Find rows that are not numeric in a Pandas DataFrame.

  Args:
    df: The Pandas DataFrame to check.

  Returns:
    A list of rows that are not numeric.
  """

  non_numeric_rows = []
  for i in range(len(df)):
    row = df.iloc[i]
    is_numeric = True
    for column in df.columns:
      if not pd.api.types.is_numeric_dtype(row[column]):
        is_numeric = False
        break
    if not is_numeric:
      non_numeric_rows.append(i)

  return non_numeric_rows

if __name__ == "__main__":


  non_numeric_rows = find_non_numeric_rows(data)
  print(non_numeric_rows)
```

```python
def visualization1(data_values,random_title):
  # Example list of numeric values

  fig = plt.figure(figsize=(20, 6))
  # Create X-axis values representing the order of data
  x_values = list(range(1, len(data_values) + 1))

  # Create the plot
  plt.plot(x_values, data_values, marker='o')

  # Set axis labels and title
  plt.xlabel('Time')
  plt.ylabel('Correlation Values')

  # Generate a random title

  plt.title(random_title)

  # Show the plot
  plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Price1": [100, 102, 104, 106, 108],
    "Price2": [105, 107, 109, 111, 113],
    "Price3": [110, 112, 114, 116, 118],
    "Price4": [115, 117, 119, 121, 123],
    "Price5": [120, 122, 124, 126, 128]
})


correlations = df.rolling(3).corr(method="pearson")

for price in correlations.columns:
    series = correlations[price]
    array = series.to_numpy()
    plt.plot(array)
    plt.title(f"Rolling Correlation for Price {price}")
    plt.show()
```
